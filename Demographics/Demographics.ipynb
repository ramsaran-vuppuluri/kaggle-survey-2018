In this notebook we are analyzing Demographics data gathered from survey takers. 

Instead of working on all the columns, we are going to extract the columns from multipleChoiceResponses.csv that are relevant to the analysis. (we are removing multipleChoiceResponses.csv data frame once the values are extracted from primary memory).

Following questions in the survey are related to Demographics:
1.	Q1--What is your gender? - Selected Choice
2.	Q1_OTHER_TEXT--What is your gender? - Prefer to self-describe - Text
3.	Q2--What is your age (# years)?
4.	Q3--In which country do you currently reside?

We can consolidate the answers as:
1.	Gender
2.	Age
3.	Location (County)


*	Q23--Approximately what percent of your time at work or school is spent actively coding?


*	Q24--How long have you been writing code to analyze data?


*	Q25--For how many years have you used machine learning methods (at work or in school)?


*	Q26--Do you consider yourself to be a data scientist?

```python
import gc

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
multipleChoiceDF = pd.read_csv('../multipleChoiceResponses.csv',low_memory=False)
```

```python
demographicsColumns = ['Q1','Q2','Q3']
```

```python
demographicsDF = multipleChoiceDF[demographicsColumns]
```

```python
multipleChoiceDF.shape[0]
```

```python
demographicsDF.info()
```

```python
del multipleChoiceDF
gc.collect()
```

```python
demographicsDF.drop([0],inplace=True)
demographicsDF.reset_index(drop=True,inplace=True)
```

```python
demographicsDF.head()
```

```python
sns.set(rc={'figure.figsize':(8,6)})
ax = sns.countplot(x=demographicsDF.Q1)

plt.xlabel("Gender")
ax.get_figure().savefig("Gender_Distribution.jpeg")
plt.show(ax);
```

```python
demographicsDF.sort_values("Q2",inplace=True)
demographicsDF.reset_index(drop=True,inplace=True)
```

```python
sns.set(rc={'figure.figsize':(12,10)})
sns.countplot(x=demographicsDF.Q2)
plt.xlabel("Age")
ax.get_figure().savefig("Age_Distribution.jpeg")
plt.show(ax);
```

```python
demographicsDF.sort_values("Q3",inplace=True)
demographicsDF.reset_index(drop=True,inplace=True)
```

```python
sns.set(rc={'figure.figsize':(20,40)})
sns.countplot(y=demographicsDF.Q3);
```

```python
demographicsDF.rename(columns={"Q1":"Gender","Q2":"Age","Q3":"Country"},inplace=True)
```

```python
demographicsDF.columns
```

```python
demographicsDF.to_csv("demographics.csv",index=False)
```

```python
del demographicsDF
gc.collect()
```

```python
experiencDF = multipleChoiceDF[['Q24','Q25']]
```

```python
experiencDF.head()
```

```python
experiencDF.drop([0],inplace=True)
experiencDF.reset_index(drop=True,inplace=True)
```

```python
experiencDF.head()
```

```python
experiencDF = pd.melt(experiencDF)
```

```python
experiencDF.head()
```

```python
experiencDF.variable.replace(to_replace='Q24',value='Coding Experience',inplace=True)
experiencDF.variable.replace(to_replace='Q25',value='Machine Leaning Experience',inplace=True)
```

```python
experiencDF.head()
```

```python
experiencDF.value.value_counts()
```

```python
experiencDF.value.replace(to_replace='I have never studied machine learning but plan to learn in the future',
                          value = 'I have never but I want to learn in the future',inplace = True)
experiencDF.value.replace(to_replace='I have never written code but I want to learn',
                          value = 'I have never but I want to learn in the future',inplace = True)
experiencDF.value.replace(to_replace='I have never studied machine learning and I do not plan to',
                          value = 'I have never and I do not plan to ',inplace = True)
experiencDF.value.replace(to_replace='I have never written code and I do not want to learn',
                          value = 'I have never and I do not plan to ',inplace = True)
```

```python
experiencDF.value.value_counts()
```

```python
experiencDF.to_csv('experienc.csv',index=False)
```
